# Boxplot, KDE

### 연속형 데이터의 분포를 사분위수 기반으로 확인

```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
import seaborn as sb
import numpy as np
```

```python
# 시작할때마다 초기화 해줘야 할 기본 조각
my_dpi = 200
font_path = 'C:\\Users\\wodyd\\Documents\\카카오톡 받은 파일./MaruBuri-Regular.ttf'
fm.fontManager.addfont(font_path)
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()
my_font_name = 'MaruBuri'
my_theme = 'dark'
sb.set_theme(style = my_theme)

plt.rcParams['font.family'] = my_font_name
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
origin = load_data('employee_data_40')
origin.head()
```

```python
width_px = 800
height_px = 350
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)
##################### 요 부분만 계속해서 바뀐다
sb.boxplot(data=origin['월수입'], orient = 'h')
#####################
# 출력
plt.grid()
plt.tight_layout()
plt.savefig('myplot.png', dpi = my_dpi)
plt.show()
plt.close()
```

### 데이터 프레임을 통한 상자그림

```python
width_px = 800
height_px = 350
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)
sb.boxplot(data = origin, x = '월수입')
ax.grid(True)
# 출력
plt.tight_layout()
plt.savefig('myplot.png', dpi = my_dpi)
plt.show()
plt.close()
```

### 복수 변수에 대한 처리

```python
width_px = 800
height_px = 350
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)
sb.boxplot(data = origin[['월수입', '나이']])
ax.grid(True)
# 출력
plt.tight_layout()
plt.savefig('myplot.png', dpi = my_dpi)
plt.show()
plt.close()
```

### KDE(커널 밀도 추정) Plot - 연속성 데이터 설정


```python
# 비모수적 방법: 정규분포를 따르는게 아닌, 데이터 기반으로 분포의 모양을 추정
# 데이터의 구조와 패턴을 더 필요로함
width_px = 1280
height_px = 640
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = my_dpi)
sb.kdeplot(data = origin['월수입'])
ax.grid(True)
# 출력
plt.tight_layout()
plt.savefig('myplot.png', dpi = my_dpi)
plt.show()
plt.close()
```
