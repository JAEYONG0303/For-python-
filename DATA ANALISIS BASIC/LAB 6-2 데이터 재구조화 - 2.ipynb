# 데이터 재구조화 - 2

```python
from hossam import load_data
from pandas import DataFrame
```

```python
origin = load_data('body_size')
origin
```

```python
st1 = origin.stack()
print(type(st1))
st1
```

### Stack -> DataFram으로 생성

```python
df1 = DataFrame(origin.stack())
df1
```

```python
# 빈 값 포함시키기
# stack -> DataFrame 전환시 자동으로 NaN(결측치)가 빠진다

st2 = origin.stack(dropna=False)
df2 = DataFrame(st2)
df2
```

### 연습문제

```python
origin = load_data('game_scores')
origin
```

```python
df1 = origin.dropna()
df2 = df1.set_index('player')
df3 = DataFrame(df2.stack())
df3
```

```python
df4 = df3.reset_index()
df4.filter(['player','0']).groupby('player').mean
df4
```
