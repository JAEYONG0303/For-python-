### 그래프 설정

```python
my_dpi = 120
my_font_name = 'MaruBuri'
my_theme = 'dark'
sb.set_theme(style = my_theme)

plt.rcParams['font.family'] = my_font_name
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
```

### 그래프 코드 기본 구성

```python
width_px = 1200
height_px = 760
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)
ax.set_title('연습용 그래프', color = '#0066ff', fontsize = 22,
            fontweight = 1000, pad = 20)
ax.set_xlabel('X축', color = '#ff6600', fontsize = 16, labelpad = 10)
ax.set_ylabel('Y축', color = '#ff00ff', fontsize = 16, labelpad = 10)

# 출력
plt.grid()
plt.tight_layout()
plt.savefig('myplot.png', dpi = my_dpi)
plt.show()
plt.close()
```

### 연습문제 - 교통사고 발생건수 시각화

```python
# 데이터 가져오기

origin = load_data('traffic_acc')
origin
```

### 데이터 전처리

```python
df = origin.drop('월', axis = 1).groupby('년도').mean()
df
```

### 교통사고 발생자 수 변화

```python
width_px = 1000
height_px = 480
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 그래프 그리기, seaborn 사용
sb.lineplot(data = df, x= df.index, y = '발생건수')


# 출력
plt.grid()
plt.tight_layout()
plt.show()
plt.close()
```

### 연습문제 - 2

```python
from pandas import pandas as pd
```

```python
origin = load_data('covid19_active')
origin
```

```python
# X좌표 직접 생성
x_index = origin.index.sort_values()
xx = list(range(0, len(x_index), 30))
xx
```

```python
xticks = []

for i in xx:
    xticks.append(x_index[i].strftime('%y/%m/%d'))

xticks
```

```python
width_px = 1260
height_px = 760
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 그래프 그리기, seaborn 사용
sb.lineplot(data = origin['서울시 일일 확진'].values, label = '서울시 일일 확진')
sb.lineplot(data = origin['전국 일일 확진'].values, label = '전국 일일 확진')

ax.grid(True)

# 문제에서 요구하는 조건
ax.set_xticks(xx, xticks, rotation = 45, fontsize = 9)
# 그래프 꾸미기

ax.set_title('코로나 변화 추이', pad = 15, fontsize = 24)
plt.grid()
plt.tight_layout()
plt.show()
plt.close()

# 객관적으로 알 수 있는 사실: 22/8월 확진자가 가장 많았다
```

### 연습문제 - 2

```python
df = load_data('bitcoin')
df
```

```python
start = list(df['시가'].values)

print(len(start))
print(start[:5])
```

```python
end = list(df['종가'].values)

print(len(end))
print(end[:5])
```

```python
x = list(range(0, len(start), 20))
print(len(x))
print(x)
```

```python
xticks = []

for i in x:
    xticks.append(df.index[i].strftime('%y/%m/%d'))
xticks[:5]
```

```python
width_px = 1800
height_px = 600
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 그래프 그리기, seaborn 사용
sb.lineplot(data = start, label = '시가')
sb.lineplot(data = end, label = '종가')

ax.grid(True)
ax.legend()
ax.set_xticks(x, xticks)
ax.set_xlim(-5, len(start))
# 문제에서 요구하는 조건
ax.set_xticks(x, xticks, rotation = 90)
# 그래프 꾸미기

ax.set_title('비트코인 변화 추이', pad = 15, fontsize = 24)
plt.grid()
plt.tight_layout()
plt.show()
plt.close()

# 객관적으로 알 수 있는 사실: 21년 6월~ 11월 말까지 상승장, 이후 하락장
```
