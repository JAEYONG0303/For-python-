### 예제 

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm # 글꼴을 시스템에 등록
import seaborn as sb
from hossam import load_data
from pandas import melt
import numpy as np
```

```python
my_dpi = 200
font_path = 'C:\\Users\\wodyd\\Documents\\카카오톡 받은 파일./MaruBuri-Regular.ttf'
fm.fontManager.addfont(font_path)
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()
my_font_name = 'MaruBuri'
my_theme = 'dark'
sb.set_theme(style = my_theme)

plt.rcParams['font.family'] = my_font_name
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
origin = load_data('unemployment_age')
origin.head()
```

```python
df = origin.copy()
df['시도별'] = df['시도별'].ffill()
df.head(10)
```

```python
df2 = melt(df, id_vars =['시도별','연령계층별'],
               var_name ='년도',
               value_name ='실업률')
df2.head(5)
```

```python
df3 = df2[['시도별','년도','실업률']].groupby(['시도별','년도'],as_index = False).mean()
df3
```

```python
df4 = df2[['년도','실업률']].groupby('년도').mean()
df4
```

### 데이터 시각화

```python
width_px = 2000
height_px = 1200
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 그래프 그리기, seaborn 사용
sb.lineplot(data = df4, x= df4.index, y = '실업률')
           

# 그래프 꾸미기
ax.grid(True) # 배경 격자 표시/숨김
ax.set_title('년도에 따른 전국 평균 실업률 변화 (전국평균)', fontsize =12, pad = 8)
# 출력
plt.tight_layout()
plt.savefig('plot.png', dpi=my_dpi * 2)
plt.show()
plt.close()
```

```python
df5 = df2.filter(['년도', '연령계층별', '실업률']).groupby(['년도', '연령계층별'], as_index = False).mean()
df5
```

```python
width_px = 2400
height_px = 1200
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 그래프 그리기, seaborn 사용
sb.lineplot(data = df5, x= '년도', y = '실업률', hue = '연령계층별')
           

# 그래프 꾸미기
ax.grid(True) # 배경 격자 표시/숨김
ax.set_title('년도에 따른 전국 평균 실업률 변화 (연령대별)', fontsize =12, pad = 8)
# 출력
plt.tight_layout()
plt.savefig('plot.png', dpi=my_dpi * 2)
plt.show()
plt.close()
```
