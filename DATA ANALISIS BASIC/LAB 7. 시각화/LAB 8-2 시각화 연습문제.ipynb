```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
import seaborn as sb
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
origin = load_data('covid19_full')
origin.head()
```

```python
df1 = origin.copy()
df1['날짜'] = pd.to_datetime(df1['날짜'])
input_date = input("날짜를 yyyymmdd형태로 입력하시오: ")
시작날짜 = datetime.strptime(input_date, "%Y%m%d")
마지막날짜 = 시작날짜 + timedelta(days=6)
조건날짜 = (df1['날짜'] >= 시작날짜) & (df1['날짜'] <= 마지막날짜)
result = df1.loc[조건날짜]
result
```

```python
df2 = result.copy()
df2 = df2.reset_index(drop = True)
df2
```

```python
df2['날짜'] = pd.to_datetime(df2['날짜'])
df2['날짜'] = df2['날짜'].dt.strftime('%m-%d')
df2
```

```python
df2 = df2.sort_values('날짜')
df2
```

```python
df3 = df2.copy()
df3['서울시 누적 확진자'] = (df2['서울시 누적 확진자']).astype(float)
df3['전국 누적 확진'] = (df2['전국 누적 확진']).astype(float)
df3['서울시 누적 확진자'] = (df2['서울시 누적 확진자'] / 1_000_000).round(3)
df3['전국 누적 확진'] = (df2['전국 누적 확진'] / 1_000_0000).round(3)
df3
```

```python
width_px = 2000
height_px = 1200
rows = 2                                                            
cols = 1                                                          
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

fig.suptitle('COVID-19', fontsize = 14, color ='#006600')

fig.subplots_adjust(wspace = 0.2, hspace = 0.2)

ax[0].set_title('서울시의 누적 확진자 및 일일 확진자 변동추이', color ='#000000', fontsize =10)
ax[0].grid()

ax[1].set_title('전국 누적 확진자 및 일일 확진자 변동추이', color ='#000000', fontsize =10)

ax[1].grid()

# 출력
plt.tight_layout()
plt.show()
plt.close()
```

```python
width_px = 2000
height_px = 2000
rows = 2                                                            
cols = 1                                                          
figsize = (width_px/ my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)
ax2 = ax[0].twinx()
fig.suptitle('COVID-19 확진자 현황', fontsize = 14, color ='#006600')
fig.subplots_adjust(wspace = 0.2, hspace = 0.2)

sb.barplot(data=df3, x='날짜', y='서울시 누적 확진자', ax=ax[0])
sb.lineplot(data=df3, x='날짜', y='서울시 일일 확진', 
    color='#FF4500',      
    linewidth=2.2,        
    marker='o',            
    markersize=7,          
    markerfacecolor='white', 
    markeredgecolor='black', 
    markeredgewidth=1.2,     
    zorder=10,ax=ax2)

ax[0].set_ylim(6.06, 6.11)
ax[0].set_yticks([6.06, 6.07, 6.08, 6.09, 6.10, 6.11])
ax[0].set_title('서울시 확진자 추이', color ='#000000', fontsize =10)
ax[0].grid()

ax2_1 = ax[1].twinx()
ax[1].set_title('전국 확진자 추이', color ='#000000', fontsize =10)

ax[1].grid()

sb.barplot(data=df3, x='날짜', y='전국 누적 확진', ax=ax[1])

# 전국 lineplot (오른쪽)
sb.lineplot(
    data=df3, x='날짜', y='전국 일일 확진', ax=ax2_1,
    color='#FF4500', linewidth=2.5, marker='o', markersize=7,
    markerfacecolor='white', markeredgecolor='black',
    markeredgewidth=1.2, zorder=10
)

# === 절대 중요한 부분 ===
# 왼쪽 y축 = 누적 확진자 (백만 단위)
ax[1].set_ylim(3.100, 3.130)
ax[1].set_yticks([3.100, 3.105, 3.110, 3.115, 3.120, 3.125, 3.130])

# 오른쪽 y축 = 일일 확진자
ax2_1.set_ylim(6000, 20000)
ax2_1.set_yticks([6000, 9000, 12000, 15000, 18000, 22000])

# --- 전국 제목 및 grid ---
ax[1].set_title("전국 확진자 추이", fontsize=11)
ax[1].grid()
plt.tight_layout()
plt.show()
plt.close() 
```
