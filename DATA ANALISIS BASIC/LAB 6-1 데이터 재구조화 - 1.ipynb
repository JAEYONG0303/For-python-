# 데이터 재구조화


```python
# Melt = 행 중심 형태로 길게 재구조화하는 변환

from hossam import load_data
from pandas import pivot_table, melt
```

### 피벗 테이블

```python
origin = load_data('city_people')
origin
```

```python
pivot1 = pivot_table(origin,
                     index = '도시',
                     columns = '연도',
                     values = '인구'
)
print(type(pivot1))
pivot1
```

```python
# 중복데이터 집계 방법 지정하기
# 인덱스와 컬럼에 따른 value가 2개 이상일 때 -> aggfunc 지정

pivot_table(origin,
            index = '지역',
            columns = '연도',
            values = '인구',
            aggfunc = 'sum'
)
```

```python
# 두 개 이상 집계 함수 지정

pivot_table(origin,
            index ='지역',
            columns = '연도',
            values = '인구',
            aggfunc = ['sum','mean']
)
```

```python
# 복수 인덱스 지정

pivot_table(origin,
            index = ['지역', '연도'],
            columns ='도시',
            values = '인구',
            aggfunc = ['sum', 'mean']
)
```

### Melt


```python
pivot = pivot_table(origin,
                    index = '연도',
                    columns = '지역',
                    values = '인구',
                    aggfunc = 'mean')
pivot
```

```python
# 데이터 프레임 인덱스를 일반 컬럼으로 설정
pivot2 = pivot.reset_index()
pivot2
```

```python
# 피벗 테이블에 대한 melt 처리

mdf = melt(pivot2, id_vars = ['연도'],
                    value_vars = ['경상권', '수도권'])
mdf
```

```python
# 이름 지정

mdf = melt(pivot2, id_vars = ['연도'],
           value_vars = ['경상권', '수도권'],
           var_name = '구분', value_name = '인구수')
mdf
```

```python
df2 = load_data('icecream_sales')
df2
```

```python
print(df2.info())
rows, cols = df2.shape
print('%d행, %d열, %d차원 데이터' % (rows, cols, df2.ndim))
```

```python
na1 = df2.dropna(subset =['Quantity']) 
na1
```

```python
pivot1 = pivot_table(na1,
                 index = 'Date',
                 columns = 'Flavor',
                 values = 'Quantity',
                 
)
pivot1
```

```python
pivot2 = pivot1.reset_index()
pivot2
```

```python
# 피벗 테이블에 대한 melt 처리

mdf = melt(pivot2, id_vars = ['Date'],
                    value_vars = ['딸기','민트초코','바닐라','초콜릿'],
                    var_name = 'flavor',
                    value_name = 'Quantity')
mdf 
```

```python
re_mdf = mdf.copy()
re_mdf = mdf.dropna(subset =['Quantity'])
re_mdf
```

```python
re_mdf1 = re_mdf.reset_index()
re_mdf1
```
