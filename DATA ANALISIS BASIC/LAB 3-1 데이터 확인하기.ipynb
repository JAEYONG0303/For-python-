# 데이터 확인하기
## 라이브러리 참조

```python
from hossam import load_data
```

```python
origin = load_data('GRADE')
origin
```

### 데이터 구조 확인 - 1

```python
# 이 과정 그대로 따라간다.

print('차원:', origin.ndim)
print('크기:', origin.shape)
```

### 데이터 구조 확인 - 2


```python
# 데이터 타입을 무조건 확인하고 가야된다.

origin.dtypes
print('-' * 50)
origin.info
```

### 데이터 내용 확인 - 1

```python
# 특정 열의 데이터 확인
print(origin['국어'])
print(type(origin['국어']))
```

### 데이터 내용 확인 - 2

```python
# 원하는 열 추출 --- 대괄호가 이중으로 표기

df2 = origin[['국어','영어','수학']]
df2
```

### 특정 행 접근 - 1

```python
# 인덱스 이름으로 접근 --loc

print(origin.loc['철수'])
print(type(origin.loc['철수']))
```

### 특정 행 접근 - 2

```python
# 인덱스 번호를 기반으로 접근 -- iloc

print(origin.iloc[0])
print('-' * 50)
print(origin[1:4])
```

### 단일 값 확인하기 - 1

```python
# 괄호의 표현에 따라 구분
# 열 -> 행 순서로 접근

origin['국어']['철수']

# 행 -> 열 순서로 접근
origin.loc['철수','국어']
```

### 단일 값 확인하기 - 2

```python
# 인덱스 번호 기반 데이터 접근
# 행 -> 열 순서로 접근

origin.iloc[1,2]
```

```python
# 인덱스 번호로 접근할 경우 슬라이싱 가능

origin.iloc[:,3]
```

### 값의 변경


```python
# 행 -> 열 순으로 접근하는 경우만 변경 가능함

origin.loc['철수','국어'] = 100
origin
```

### 데이터 일부만 확인

```python
# 상위 n건 확인
# 파라미터 생략될 경우 기본값 5

origin.head(2)
```

```python
# 하위 n건 확인

origin.tail(2)

# 데이터 프레임의 크기가 매우 클 것이므로, 일부 추출 후 형태를 가늠해야한다
```
