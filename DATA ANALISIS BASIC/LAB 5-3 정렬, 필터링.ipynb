# 정렬, 필터링

```python
from hossam import load_data
from pandas import DataFrame #query 함수 
```

```python
origin = load_data('grade')
origin
```

```python
# drop vs filter는 나에게 어떤 데이터가 더 필요하냐에 따라서 쓴다. 상황봐서 잘 써야 한다.
```

### 데이터 정렬

```python
# 변수에 의한 정렬
# ascending = False 파라미터는 내림차순 정렬을 설정함(기본값 = True, 오름차순)

단일정렬df = origin.sort_values('학년', ascending = False)
단일정렬df
```

```python
# 여러 변수 정렬

복수정렬df = origin.sort_values(['학년', '국어'],
                            ascending = [False, False])
복수정렬df
```

```python
# 인덱스 기준 정렬

인덱스정렬df = origin.sort_index(ascending = False)
인덱스정렬df
```

### 데이터 필터링

```python
# 행 단위 필터링(조건검색용도)
# 숫자 타입
r1 = origin.query('국어 > 80')
r1
```

```python
# 행 단위 필터링(조건검색용도)
# 문자 타입

r2 = origin.query("성별 == '남자'")
r2
```

```python
# 행 단위 필터링
# AND 조건

r3 = origin.query('국어 > 80 and 수학 > 80')
r3
```

```python
# 행 단위 필터링
# OR 조건

r4 = origin.query('국어 <= 80 or 수학 <= 80')
r4
```

```python
# 행 단위 필터링
# 조건값을 변수로 적용하기

point = int(input('점수 입력하시오:'))
print('국어 점수가 %d점 보다 높은 학생' % point)

r5 = origin.query('국어 > @point')
r5
```

```python
# 열 단위 필터링
fl = origin.filter(['국어', '영어'])
fl
```
