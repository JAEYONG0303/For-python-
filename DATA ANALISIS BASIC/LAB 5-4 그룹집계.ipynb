# 그룹집계


```python
from hossam import load_data
```

```python
origin = load_data('traffic_acc')
origin
```

### 지정된 컬럼을 기준으로 그룹화

```python
# 전체 컬럼에 대한 집계

origin.groupby('년도').sum()
```

```python
# 불필요한 컬럼을 제외하는 경우

df = origin.drop(columns = '월')
df
```

```python
# 새로운 데이터 프레임 활용한 데이터 집계

df.groupby('년도').sum()
```

### 필요한 컬럼만 추출하여 집계하는 경우

```python
# 년도와 교통사고 사망자수만 추출

df = df.filter(['년도', '사망자수'])
df
```

```python
# 추출된 새로운 데이터 프레임을 활용한 집계

df.groupby('년도').sum()
```

### 두 개 이상의 번수에 대한 그룹화

```python
# 2개 이상 그룹화 할 경우
# 리스트로 묶어서 설정한다.

origin.groupby(['년도','월']).mean()
```

```python
# 컬럼별로 집계함수를 다르게 지정하는 경우
# agg()함수 + {컬럼:함수명} 딕셔너리 전달

origin.drop('월', axis=1).groupby('년도').agg(
    {'발생건수': ['mean', 'sum'], '사망자수': 'max', '부상자수': 'min'}
)
```

### 사용자 정의 함수를 사용한 집계

```python
# 사용자 정의 함수 적용

def myfunction(x):
    # x는 Series객체 - 컬럼 하나가 통쨰로 전달됨
    return x.max() - x.min()
# 사용자 정의 함수 이름은 문자열이 아니다.
origin.drop('월', axis = 1).groupby('년도').agg(['max', 'min', myfunction])
```

### 인덱스 지정 해제

```python
# groupby 매서드에서 사용한 필드는 결과집합의 인덱스로 지정되는 것이 기본 형태
# as_index = False 사용으로 인덱스 지정 방지 가능
origin.drop('월', axis = 1).groupby('년도', as_index = False).sum()
```
