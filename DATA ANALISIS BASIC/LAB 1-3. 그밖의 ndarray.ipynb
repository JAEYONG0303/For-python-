# 그 밖의 ndarray 형태
## 라이브러리 참조하기

```python
import numpy as np
```

### 2차원 배열

```python
# 2차원 리스트를 ndarray로 변환한다

myarray = np.array([
    [98, 72, 80, 64],
    [88, 90, 80, 72],
    [92, 88, 82, 76]
])

print(myarray)

print('차원크기:', myarray.ndim)
print('차원별 원소수', myarray.shape)
print('원소의 데이터 타입', myarray.dtype)
```

### boolean 타입으로 인덱싱

```python
# 조건에 맞는 원소를 '1차원 배열'로 반환

bool_array = np.array([
    [True, False, True, False],
    [True, True, True, False],
    [True, True, True, False]
])

# 조건에 맞는 항목(True 값)을 1차 배열로 추출
result = myarray[bool_array]
result
```

### 정형화된 배열 생성

```python
# 모든 원소가 0인 배열
print(np.zeros([3,4]))
print('-'*50)
# 모든 원소가 1인 배열
print(np.ones([3,4]))
print('-'*50)
# 모든 원소가 특정 값인 배열
print(np.full([3,4], 10))
print('-'*50)
# 대각선 원소가 1, 그 외에는 0인 배열
print(np.eye(4))
```

### 랜덤 배열(1)


```python
# 0~1 범위에서 균등하게 추출하여 N차원 랜덤 배열 생성
# 10개의 값을 가지는 1차원 배열
#np.random.rand(10)

# 4행 3열의 값을 가지는 1차원 랜덤 배열
np.random.rand(4,3)
```

### 주어진 범위 안에서 정수값을 갖는 랜덤 배열


```python
# 0~9사이 랜덤값 10개의 원소를 갖는 1차원 배열 --> 중복 존재

# np.random.rand(0,9,10) --> 이렇게 값 입력시 다르게 나옴!
np.random.randint(0, 9, 10)
# np.random.randfloat(0,9,10) --> 존재하지 않는 모듈이라 나옴

# 0~9사이 랜덤값 10개의 원소를 갖는 3행 4열의 2차원 배열
np.random.randint(0,9,(3,4))
```

### 랜덤 시드 고정


```python
# 랜덤 시드

np.random.seed(1234)
np.random.normal(0,1,(3,4))
```

### 표준 정규 분포를 따르는 랜덤 배열

```python
# 원소가 10개인 1차원 배열
# 이 매서드 함수 자체가 평균 = 0, 분산 = 1 을 따른다는 가정이 들어간 상태.
np.random.randn(10)

# 원소가 12개인 4행 3열의 2차원 배열

np.random.randn(4,3)
```
