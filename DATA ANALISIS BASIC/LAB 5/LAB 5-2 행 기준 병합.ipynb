# 행 기준 병합


```python
from pandas import DataFrame
from pandas import concat
```

```python
df1 = DataFrame({'이름': ['영희', '철수'], '국어': [87, 91]})
print(df1)

df2 = DataFrame({'이름': ['민철', '수현'], '국어': [78, 92]})
print(df2)
```

### concat() 함수 기본 사용 방법


```python
# 병합하기

concat([df1, df2])
```

```python
# 병합된 결과에서 index 재구성

인덱스_재구성df = concat([df1, df2])
인덱스_재구성df.reset_index(inplace =True, drop =True)
인덱스_재구성df
```

```python
# 병합 후 ignore_index =True 사용하여 인덱스 재구성

concat([df1, df2], ignore_index = True)
```

### 서로 다른 변수를 갖는 데이터 프레임 병합

```python
df3 = DataFrame({'이름': ['영희', '철수'], '국어': [87, 91]})
print(df3)
print(df3.info())
df4 = DataFrame({'이름': ['민철', '수현'], '수학': [78, 92]})
print(df4)
```

```python
# 데이터 병합
concat([df3,df4], ignore_index = True)
```

### 인덱스를 갖는 데이터 프레임 병합

```python
df5 = DataFrame({'국어': [87, 91, 85]},
                index = ['영희','철수','민수'])
print(df5)

df6 = DataFrame({'국어': [78, 92, 82]},
                index = ['민철','수현','민수'])
print(df6)
```

```python
# 데이터 병합 진행하면, 인덱스도 함께 병합
# 증복되는 인덱스 존재할 경우, 각 행은 개별적 존재(인덱스는 중복된 채로)

concat([df5,df6])
```

```python
# 인덱스 제외하고 병합하기
# 정확도가 떨어진다
concat([df5, df6], ignore_index =True)
```
