# DATA PREPROCESSING
## 데이터 구조 점검 - 1

```python
# 데이터 전처리

from hossam import load_data
from pandas import CategoricalDtype # (명목형 타입을 가져오는 클래스)
```

```python
origin = load_data('grade')
origin
```

### 데이터 확인하기

```python
# 데이터 크기

rows, cols = origin.shape
print('%d행, %d열의 %d차원 데이터' % (rows, cols, origin.ndim))
```

### 데이터 타입 변경

```python
# 컬럼 단위 변경
# 작업을 위한 복사본 생성(아주 중요한 개념)

컬럼타입변경df = origin.copy()
컬럼타입변경df.info()
```

```python
# 컬럼 단위 변경

컬럼타입변경df['국어'] = 컬럼타입변경df['국어'].astype('float')
컬럼타입변경df.info()
```

```python
# 타입 일괄 변경 후 DataFrame 리턴 받기

타입변환df = origin.astype({'국어':'float', '영어': 'float',
                            '수학': 'str', '과학': 'bool'})
타입변환df.info()
```

```python
# 타입 일괄 변경 후 원본 확인

origin.info()
```

```python
# 범주형 데이터를 명목척도로 변환
# 컬럼 단위 변환

컬럼단위_명목df = origin.copy()
컬럼단위_명목df['학년'] = 컬럼단위_명목df['학년'].astype('category')
컬럼단위_명목df['성별'] = 컬럼단위_명목df['성별'].astype('category')
컬럼단위_명목df.info()
```

```python
# 범주형 데이터를 명목척도로 변환
# 복사본 리턴 받기

범주형df = origin.astype({'학년': 'category', '성별': 'category'})
범주형df.info()
```

```python
# 범주형 데이터를 서열척도로 변환
# 컬럼 단위로 적용하여 원본에 직접 적용하기

컬럼단위_서열척도df = origin.copy()
컬럼단위_서열척도df['학년'] = 컬럼단위_서열척도df['학년'].astype(
    CategoricalDtype(categories = [1,2,3,4], ordered = True))
컬럼단위_서열척도df.info()
```

```python
# 서열척도로 변환 후 복사본 리턴 
서열척도df = origin.astype({
    '학년': CategoricalDtype(categories = [1, 2, 3, 4], ordered = True)})
서열척도df.info()
```

### 행, 열의 이름 변경

```python
# rename() 매서드의 이용
# 기존이름: 딕셔너리로 지정
# columns와 index 중 변경을 원하는 파라미터 하나만 설정 가능
# 원본 변화 없음.
# 이름 변경되지 않은 항목이 결과에서 제외되지는 않는다
이름변경df = origin.rename(
    columns = {'국어': 'kor', '영어': 'eng', '수학': 'math', '과학': 'sci'},
    index = {'석영': 'suk', '호영': 'ho', '민철': 'min', '수현': 'su'}
)
이름변경df
```
