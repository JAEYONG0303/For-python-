```python
# 라이브러리 참조
import numpy as np
```

### 샘플 데이터 준비

```python
# 학급 시험 점수
point = np.array([100, 91, 89, 86, 84, 79, 78, 77, 74, 71, 69, 66, 65, 60, 58, 57, 55])
point
```

### Numpy의 histogram: 도수 분포표 - 1


```python
# 5개 구간으로 나누기
hist, bins = np.histogram(point, 5)
print('구간별 데이터 수:', hist)
print('구간 경계:', bins)
```

### Numpy의 histogram: 도수 분포표 - 2


```python
# 도수 분포표 구성
s = len(hist)
for i in range(0, s):
    if i+1 <s:
        tpl = '%d이상 %d미만: %d개'
    else:
        tpl = '%d이상 %d이하: %d개'
    print(tpl % (bins[i], bins[i+1], hist[i]))
```

### Numpy의 histogram: 도수 분포표 - 3

```python
# 데이터 구간 직접 정의하기

hist, bins = np.histogram(point, [50,60,70,80,90,100])
print('구간별 데이터 수:', hist)
print('구간 경계:', bins)
```

### 도수의 종류


```python
# 상대도수
relative_freq = hist / np.sum(hist)
relative_freq
```

```python
# 누적 도수

Sum_freq = []

previous = 0

for i in hist:
    previous += i
    Sum_freq.append(previous)

previous
```

### 연습문제 - 2


```python
# 중앙값 계산

Crash = np.array([15129, 13836, 15738, 16300, 17605,16690, 16252, 17000, 16956, 17720, 17455, 15668])
print(Crash)

# 데이터 크기 계산

size = len(Crash)
print('데이터 크기:', size)

# 데이터 정렬
sorted_Crash = np.sort(Crash)
print('데이터 정렬:', sorted_Crash)

# 짝수인 경우..

if size % 2 == 0:
    a = (size // 2) -1
    b = a + 1
    median = (sorted_Crash[a] + sorted_Crash[b]) / 2
else:
    a = (size // 2)
    median = a
print('중앙값:', median)
```

```python
# 1 사분위수 계산

#Crash = np.array([15129, 13836, 15738, 16300, 17605,16690, 16252, 17000, 16956, 17720, 17455, 15668])
#print(Crash)

# 방법 1. 직접 계산
a = np.sort(Crash)
print(np.sort(Crash))
size = len(a) 
q1_pos = (size-1) * 0.25
b = int(np.floor(q1_pos))
f = q1_pos - b

# 보간
Q1 = round(a[b] + f *(a[b+1] - a[b]),0)
print('직접 계산 Q1:', Q1)

# 내장 함수 사용
print('내장 함수 사용:', round(np.quantile(a, 0.25),0))
```

```python
# 3 사분위수 계산

# 방법 1. 직접 계산
a = np.sort(Crash)
print(np.sort(Crash))
size = len(a) 
q3_pos = (size-1) * 0.75
b = int(np.floor(q3_pos))
f = q3_pos - b

# 보간
Q3 = round(a[b] + f *(a[b+1] - a[b]),0)
print('직접 계산 Q3:', Q3)

# 내장 함수 사용
print('내장 함수 사용:', round(np.quantile(a, 0.75),0))

# 표준편차
print('표준편차:', round(np.std(Crash)))
print('분산:', np.var(Crash))
```

```python
# 절대도수, 상대도수, 누적도수

# 절대도수

hist, bins = np.histogram(Crash, [13000,14000,15000,16000,17000,18000])
print('구간별 데이터 수:', hist)
print('구간 경계:', bins)

# 상대도수

relative_freq = hist / np.sum(hist)
print(relative_freq)

# 누적도수

cum_freq = []

previous = 0

for i in hist:
    previous += i
    cum_freq.append(previous)

previous
```

```python
# 정리

hist, bins = np.histogram(Crash, [13000,14000,15000,16000,17000,18000])
print('구간별 데이터 수:', hist)
print('구간 경계:', bins)
print('3사분위수:',Q3)
print('1사분위수:',Q1)
print('4사분위수:', max(Crash))
print('사분위수 구간(IQR):', round((np.quantile(Crash, 0.7) - np.quantile(Crash, 0.25)), 0))
print('범위:', max(Crash) - min(Crash))
print('중앙값:', median)
print('표준편차:', round(np.std(Crash)))
print('분산:', round(np.var(Crash)))
```

### 연습문제 -2 해설

구간별 데이터 수를 살펴보았을 때, 교통사고 건수는 연중 일정하게 유지되며 대부분의 달이 15,000~18,000 범위에 속한다.
즉, IQR 값과 표준편차를 통해 전체적으로 안정적인 분포를 보이는 것을 알수있다.
제공된 데이터 값을 살펴보면, 2월 교통사고 건수가 제일 적고 10월 교통사고 건수가 제일 많은 점을 볼 수 있다.
이와 같은 현상을 해석하고자 환경적 상황에 대한 이해가 필요하다.
2월의 경우 전체 12월에서 가장 적은 일 수(28일)를 가지고 있으며, 본격적으로 일몰 시각이 늦어지기 시작한다.
10월의 경우 본격적으로 일몰 시간이 빨라지기 시작한다.
