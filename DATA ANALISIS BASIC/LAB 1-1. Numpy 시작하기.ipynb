### 배열 생성하기 연습문제

```python
# 1번 문제
arr = np.array([10, 20, 30, 40, 50])
print(type(arr))
size = len(arr)
arr
print("배열의 길이: %d"% size)
```

```python
# 2번 문제
print("첫 번째 원소:",arr[0])
print("마지막 두 개의 원소:",arr[-2:])
```

```python
# 3번 문제
for i,v in enumerate(arr):
    print("%d번째 원소는 --> %s" %(i,v))
```
