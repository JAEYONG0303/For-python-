### 데이터 불러오기

```python
# 과목별 점수 데이터
from hossam import load_data
origin = load_data('grade')
```

### 엑셀로 저장하기 - 1


```python
# 파일 이름만 지정

origin.to_excel('1.xlsx')
```

### 엑셀로 저장하기 - 2

```python
# 파라미터 설정 - 인덱스를 포함하는 경우

origin.to_excel('2.xlsx',
                index = True,
                index_label = 'name')
```

### 엑셀로 저장하기 - 3


```python
# 파라미터 설정 - 인덱스 미포함 일때

origin.to_excel('3.xlsx',
                index = False,
                columns = ['학년', '성별', '국어'],
                header = ['level', 'sex','kor'])
```

### CSV로 내보내기 - 1

```python
# 원본 그대로 저장하기
origin.to_csv('1.csv', encoding = 'utf - 8')
```

### CSV로 내보내기 - 2

```python
# 파라미터 설정 - 인덱스 설정 하는 경우

origin.to_csv('2.csv',
              encoding = 'utf-8',
              index = True,
              index_label = 'Name')
```

### CSV로 내보내기 - 3

```python
# 파라미터 설정 - 인덱스 포함하지 않으면서 컬럼 이름 직접 설정

origin.to_csv('3.csv',
              encoding = 'utf-8',
              index = False,
              columns = ['학년', '성별','국어'],
              header = ['level', 'sex','kor'])
```
